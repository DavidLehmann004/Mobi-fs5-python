```python
import numpy as np
import matplotlib.pyplot as plt
```

**[Task 1.2 (a).]()** Use `plt.imread` to load an image:

```python
img = plt.imread("data/NIH3T3/im/dna-0.png")
```

**[Task 1.2 (b).]()** Show the image:

```python
plt.imshow(image, cmap='gray')
plt.colorbar()
```

**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
tresh = 0.1
binary = img >= tresh
plt.figure()
plt.imshow(binary, cmap='gray')
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
gt_float = plt.imread("data/NIH3T3/gt/0.png")
plt.imshow(gt_float, cmap='gray')
gt_bool = gt_float.astype(bool)
```

Compute and print the Dice coefficient:

```python
dice_score = 2 * (binary*gt_bool).sum()/(binary.sum() + gt_bool.sum())
dice_score
```
